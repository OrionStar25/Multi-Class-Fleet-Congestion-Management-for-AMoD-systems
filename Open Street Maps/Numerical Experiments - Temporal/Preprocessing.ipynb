## Calculate congestion of network for off-peak

```python
G = ox.save_load.load_graphml(filename="Singapore_drive.graphml")
edges = ox.graph_to_gdfs(G, nodes=False, edges=True)

print(type(G))

f = open("data/traffic/Sun_12am_off.json", "r").read()
x = json.loads(f)
speed_bands = pd.DataFrame.from_dict(x, orient='columns')
process_speed_band(speed_bands)
calculate_congestion(G, edges, speed_bands)

edges.head()
```

```python
make_net_tntp(edges, 'data/Singapore_net_off.tntp')
```

```python
ox.save_load.save_graph_shapefile(G, filename='off')
```

# Make demand file (1000 customers)

```python
nodes_central = pd.read_pickle('data/Central_nodes.pkl')
nodes_central.drop([7330643916], inplace=True)
```

```python
# get 20 sources from central SG
sources = nodes_central['osmid'].values[:25]
 
# get 15 targets from central SG
targets = nodes_central['osmid'].values[-18:]
```

```python
demands = []
for s in sources:
    for t in targets:
        demands.append([s,t])
```

```python
make_trips_tntp(demands)
```
