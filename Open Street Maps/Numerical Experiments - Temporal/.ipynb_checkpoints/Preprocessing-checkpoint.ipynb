# Make demand file (500 customers)

```python
import osmnx as ox
import pandas as pd
import pickle
import networkx as nx

from rtree import index
from utils import *
```

```python
nodes_central = pd.read_pickle('data/Central_nodes.pkl')
nodes_central.drop([7330643916], inplace=True)
```

```python
# get 10 sources from central SG
sources = nodes_central['osmid'].values[51:61]
 
# get 10 targets from central SG
targets = nodes_central['osmid'].values[-61:-51]
```

```python
G = ox.save_load.load_graphml(filename="Singapore_drive.graphml")
nodes = ox.graph_to_gdfs(G, nodes=True, edges=False)
```

```python
# Fill in all nodes to find neighbours
# key: node label ,value: location
idx = index.Index()

xx = nodes['x'].astype(float)
yy = nodes['y'].astype(float)
node_id = nodes['osmid']

for node, x, y in zip(node_id, xx, yy):
    idx.insert(int(node), (x,y,x,y))
```

```python
demands = []
for s in sources:
    for t in targets:
        demands.append([s,t])
```

```python
make_trips_tntp(demands, "data/Singapore_trips_500_4.tntp")
```

## Make demand file for middle mile (500 customers)

```python
def bingo(sources, targets): 
    demand = []    
    count = 0
    
    for s in sources:
        for t in targets:
            orig_point = nodes[nodes['osmid'] == int(s)].geometry.values[0]
            target_point = nodes[nodes['osmid'] == int(t)].geometry.values[0]
            orig_xy = (orig_point.y, orig_point.x)
            target_xy = (target_point.y, target_point.x)

            # Find all nodes within 500m from candidate sources and destinations
            candidate_dest = find_nearest_nodes(nodes, target_xy, 500, idx)
            candidate_source = find_nearest_nodes(nodes, orig_xy, 500, idx)

            euclidean = get_length_dict(nodes, candidate_source, candidate_dest)        
            OD = modified_hybrid_search(nodes, euclidean)
            demand.append(OD)
            count += 1
            print(count)
            
    return demand
```

```python
G_ped = ox.save_load.load_graphml(filename="Singapore_walk.graphml")
```

### Off-peak 

```python
transits_off = bingo(sources, targets)
```

```python
ped_off = []
ctr = 0

for s in sources:
    for t in targets:
        first = nx.dijkstra_path_length(G, s, transits_off[ctr][0], weight='length')/3.7
        last = nx.dijkstra_path_length(G, transits_off[ctr][1], t, weight='length')/3.7
        ped_off.append(first+last)
```

```python
with open('data/ped_off_4.pkl', 'wb') as fp:
    pickle.dump(ped_off, fp)
```

```python
make_trips_tntp(demands, filename="data/Singapore_trips_500_off_4.tntp")
```
